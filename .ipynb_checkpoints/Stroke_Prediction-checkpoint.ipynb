### Background

---
Stroke is a leading cause of death worldwide, imposing substantial health and economic burdens on patients and healthcare systems. Health-related behaviors are recognized as key risk factors for stroke and are increasingly targeted for prevention efforts. While numerous machine learning models have been developed to predict stroke risk or assist in diagnosis using lifestyle factors or medical imaging, there is a lack of models utilizing laboratory test data for stroke prediction.

---

### Objective

---

The objective of this study is to apply machine learning methods to laboratory test data in order to build predictive models that can accurately identify individuals at risk of stroke.

### Feature Engineering
---

```python
##Imputing bmi for improving the model Accuracy and Recall

features_for_imputation = ['age', 'avg_glucose_level', 'bmi']


impute_df = df[features_for_imputation].copy()


knn_imputer = KNNImputer(n_neighbors=5)
imputed_array = knn_imputer.fit_transform(impute_df)


df['bmi'] = imputed_array[:, impute_df.columns.get_loc('bmi')]
```

```python
# Map smoking_status to risk scores
smoking_risk_map = {
    'never smoked': 0,
    'formerly smoked': 1,
    'smokes': 2,
    'Unknown': 1
}

df['smoking_risk'] = df['smoking_status'].map(smoking_risk_map).fillna(0).astype(int)

# Adding combined risk feature
df['cardiovascular_risk'] = df['hypertension'] + df['heart_disease'] + df['smoking_risk']


print(df['cardiovascular_risk'].value_counts().sort_index())
print(df['smoking_risk'].value_counts().sort_index())
```

---
### Class Imbalance Effects

The model achieves a high recall for the positive class (around 79%), meaning it detects most of the stroke cases. However, the precision for this class remains low (about 50%), indicating many false positives.

This imbalance between recall and precision happens because the model struggles to be precise despite detecting most positive cases. SMOTE helps by generating synthetic samples for the minority class, but the challenge of imbalanced data still affects model performance.

Overall accuracy seems good (~76%), but the differing precision and recall values highlight the need for careful handling and tuning when working with imbalanced datasets.

### Using SMOTE on Logistic Regression 
---

```python
categorical_features = ['gender', 'ever_married', 'work_type', 'Residence_type', 'smoking_status', 'hypertension', 'heart_disease','bmi_category']
numeric_features = ['age', 'avg_glucose_level', 'bmi' ,'cardiovascular_risk','smoking_risk']

X = df[categorical_features + numeric_features]
y = df['stroke']

X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                    test_size=0.2,
                                                    stratify=y,
                                                    random_state=42)

preprocessor = ColumnTransformer(
    transformers=[
        ('cat', Pipeline([
            ('ohe', OneHotEncoder(handle_unknown='ignore')),
            ('scaler', StandardScaler(with_mean=False))  # sparse + scaler
        ]), categorical_features),
        
        ('num', StandardScaler(), numeric_features)
    ]
)

# Preprocess the data
X_train_processed = preprocessor.fit_transform(X_train)
X_test_processed = preprocessor.transform(X_test)




# Apply SMOTE on training data only

#bsmote = BorderlineSMOTE(random_state=42, kind='borderline-1')
#X_train_resampled, y_train_resampled = bsmote.fit_resample(X_train_processed, y_train)

#adasyn = ADASYN(random_state=42)
#X_train_resampled, y_train_resampled = adasyn.fit_resample(X_train_processed, y_train)

smote = SMOTE(random_state=42)
X_train_resampled, y_train_resampled = smote.fit_resample(X_train_processed, y_train)

print(f"Original training set: {X_train_processed.shape}")
print(f"Resampled training set: {X_train_resampled.shape}")
print(f"Original class distribution: {np.bincount(y_train)}")
print(f"Resampled class distribution: {np.bincount(y_train_resampled)}")

# Train Logistic Regression on resampled data
model = LogisticRegression(class_weight='balanced', max_iter=1000)
model.fit(X_train_resampled, y_train_resampled)

# Predict probabilities and classes
y_pred_proba = model.predict_proba(X_test_processed)[:, 1]
y_pred_class = (y_pred_proba >= 0.5).astype(int)

# Evaluate
accuracy = accuracy_score(y_test, y_pred_class)
roc_auc = roc_auc_score(y_test, y_pred_proba)
print(f"\n=== Model Performance ===")


print(f"Accuracy: {accuracy:.4f}")
print(f"ROC-AUC: {roc_auc:.4f}")
print("Classification Report (Logistic Regression with SMOTE):")
print(classification_report(y_test, y_pred_class))

# Plotting
report_dict = classification_report(y_test, y_pred_class, output_dict=True)
report_df = pd.DataFrame(report_dict).transpose()
report_df['model'] = 'Logistic Regression with SMOTE'   

all_reports_df = pd.concat([all_reports_df, report_df.reset_index()], ignore_index=True)

cm = confusion_matrix(y_test, y_pred_class)

fig, axes = plt.subplots(1, 2, figsize=(16, 6))

sns.heatmap(report_df.iloc[:-1, :-1], annot=True, cmap='Blues', fmt='.2f',
            ax=axes[0], linewidths=0.5, linecolor='gray')
axes[0].set_title("Classification Report (Logistic Regression with SMOTE)", fontsize=14, fontweight='bold')
axes[0].tick_params(axis='y', rotation=0)

disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=model.classes_)
disp.plot(cmap='Blues', ax=axes[1], colorbar=False)
axes[1].set_title("Confusion Matrix (Logistic Regression with SMOTE)", fontsize=14, fontweight='bold')
axes[1].grid(False)


plt.tight_layout()
plt.savefig("data/classification_report_and_confusion_matrix_Logistic_Regression_SMOTE.png", dpi=300)
plt.show()
```

---
###  Threshold tuning as SMOTE alone did not work
---

```python
#warnings.filterwarnings("ignore", message=".*sklearn.utils.parallel.delayed.*")

y_proba_lr_smote = model.predict_proba(X_test_processed)[:, 1]

precision_lr_smote, recall_lr_smote, thresholds_lr_smote = precision_recall_curve(y_test, y_proba_lr_smote)

# Compute F1 scores for all thresholds  
f1_scores_lr_smote = 2 * (precision_lr_smote[:-1] * recall_lr_smote[:-1]) / (precision_lr_smote[:-1] + recall_lr_smote[:-1] + 1e-10)

best_index_lr_smote = np.argmax(f1_scores_lr_smote)
best_threshold_lr_smote = thresholds_lr_smote[best_index_lr_smote]
best_f1_lr_smote = f1_scores_lr_smote[best_index_lr_smote]

print(f"Best threshold (Logistic Regression with SMOTE): {best_threshold_lr_smote:.4f}")
print(f"Best F1 score (Logistic Regression with SMOTE): {best_f1_lr_smote:.4f}")

# Apply the best threshold
y_pred_opt_lr_smote = (y_proba_lr_smote >= best_threshold_lr_smote).astype(int)
print("\nClassification Report (Logistic Regression with SMOTE and Optimized Threshold):")
print(classification_report(y_test, y_pred_opt_lr_smote, digits=2))

# Plot F1 score vs Threshold
plt.figure(figsize=(8,6))
plt.plot(thresholds_lr_smote, f1_scores_lr_smote, label="F1 Score")
plt.axvline(best_threshold_lr_smote, color='r', linestyle='--', label=f'Best Threshold: {best_threshold_lr_smote:.2f}')
plt.xlabel("Threshold")
plt.ylabel("F1 Score")
plt.title("F1 Score vs Threshold (Logistic Regression with SMOTE)")
plt.legend()
plt.grid(True)
plt.show()

# Hyperparameter tuning for Logistic Regression with SMOTE
print("\nHyperparameter Tuning for Logistic Regression with SMOTE:")
lr = LogisticRegression(max_iter=2000, random_state=42)

param_dist_lr = [
    {'solver': ['liblinear'], 
     'penalty': ['l1', 'l2'], 'C': [0.001, 0.01, 0.1, 1, 10, 100],
     'class_weight': [None, 'balanced', {0: 1, 1: 2}, {0: 1, 1: 5}, {0: 1, 1: 20}]},
]


random_search_lr = RandomizedSearchCV(
    estimator=lr,
    param_distributions=param_dist_lr,
    n_iter=50,  
    scoring=make_scorer(recall_score),  
    cv=StratifiedKFold(n_splits=5),
    random_state=42,
    n_jobs=-1,  
)

# Fit on the resampled training data (after SMOTE)
random_search_lr.fit(X_train_resampled, y_train_resampled)

print("Best Parameters (Logistic Regression with SMOTE):", random_search_lr.best_params_)

best_lr_smote = random_search_lr.best_estimator_

# Evaluate the best model
y_pred_proba_tuned_lr = best_lr_smote.predict_proba(X_test_processed)[:, 1]
y_pred_class_tuned_lr = (y_pred_proba_tuned_lr >= 0.5).astype(int) # Using default threshold initially

print("\nClassification Report (Tuned Logistic Regression with SMOTE):")
print(classification_report(y_test, y_pred_class_tuned_lr, digits=2))

# Optional: Further threshold tuning after hyperparameter tuning
precision_tuned_lr, recall_tuned_lr, thresholds_tuned_lr = precision_recall_curve(y_test, y_pred_proba_tuned_lr)
f1_scores_tuned_lr = 2 * (precision_tuned_lr[:-1] * recall_tuned_lr[:-1]) / (precision_tuned_lr[:-1] + recall_tuned_lr[:-1] + 1e-10)

best_index_tuned_lr = np.argmax(f1_scores_tuned_lr)
best_threshold_tuned_lr = thresholds_tuned_lr[best_index_tuned_lr]

print(f"\nBest threshold (Tuned Logistic Regression with SMOTE): {best_threshold_tuned_lr:.4f}")

y_pred_opt_tuned_lr = (y_pred_proba_tuned_lr >= best_threshold_tuned_lr).astype(int)
print("\nClassification Report (Tuned Logistic Regression with SMOTE and Optimized Threshold):")
print(classification_report(y_test, y_pred_opt_tuned_lr, digits=2))

report_dict = classification_report(y_test, y_pred_opt_tuned_lr, digits=2,output_dict=True)
report_df = pd.DataFrame(report_dict).transpose()
report_df['model'] = 'Logistic Regression with SMOTE,Threashold & Hyperparameter Tuning'   
all_reports_df = pd.concat([all_reports_df, report_df.reset_index()], ignore_index=True)

cm = confusion_matrix(y_test, y_pred_opt_tuned_lr)

fig, axes = plt.subplots(1, 2, figsize=(16, 6))

sns.heatmap(report_df.iloc[:-1, :-1], annot=True, cmap='Blues', fmt='.2f',
            ax=axes[0], linewidths=0.5, linecolor='gray')
axes[0].set_title("Classification Report (Logistic Regression with SMOTE Tuned)", fontsize=14, fontweight='bold')
axes[0].tick_params(axis='y', rotation=0)

disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=model.classes_)
disp.plot(cmap='Blues', ax=axes[1], colorbar=False)
axes[1].set_title("Confusion Matrix (Logistic Regression with SMOTE Tuned)", fontsize=14, fontweight='bold')
axes[1].grid(False)


plt.tight_layout()
plt.savefig("data/classification_report_and_confusion_matrix_Logistic_Regression_Tuned_SMOTE.png", dpi=300)
plt.show()
```

---
### Conclusion on Logistic Regression Performance

**Strengths**  
- Balanced recall (~78-79%) for stroke cases after SMOTE and tuning, helping detect most actual positive cases.

**Weaknesses**  
- Precision remains low (~50%), meaning many false positives that could cause unnecessary follow-ups.  
- Accuracy (~76-77%) is misleading due to class imbalance, so it doesn't fully reflect model effectiveness.  
- High classification threshold (~0.94) after tuning indicates the model needs stricter decision boundary to balance metrics.

### GradientBoosting Classifier
---

```python
# Define feature groups
categorical_features = ['gender', 'ever_married', 'work_type', 'Residence_type',
                        'hypertension', 'smoking_status', 'heart_disease', 'bmi_category']
numeric_features = ['age', 'avg_glucose_level', 'bmi', 'cardiovascular_risk', 'smoking_risk']

# Split data
X = df[categorical_features + numeric_features]
y = df['stroke']
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)

# Preprocessing pipeline
preprocessor = ColumnTransformer(
    transformers=[
        ('cat', Pipeline([
            ('ohe', OneHotEncoder(handle_unknown='ignore')),
            ('scaler', StandardScaler(with_mean=False))  # Because OneHot produces sparse
        ]), categorical_features),
        ('num', StandardScaler(), numeric_features)
    ]
)

# Fit and transform training data, transform test data
X_train_processed = preprocessor.fit_transform(X_train)
X_test_processed = preprocessor.transform(X_test)

# Apply SMOTE
smote = SMOTE(random_state=42)
X_train_resampled, y_train_resampled = smote.fit_resample(X_train_processed, y_train)

# Train Gradient Boosting model
model = GradientBoostingClassifier(n_estimators=100, learning_rate=0.1, max_depth=3, random_state=42)
model.fit(X_train_resampled, y_train_resampled)

# Predict class probabilities
y_pred_proba = model.predict_proba(X_test_processed)[:, 1]
y_pred_class = (y_pred_proba >= 0.5).astype(int)


print("Classification Report:")
print(classification_report(y_test, y_pred_class))

# Create report DataFrame
report_dict = classification_report(y_test, y_pred_class, output_dict=True)
report_df = pd.DataFrame(report_dict).transpose()

report_df['model'] = 'GradientBoosting with SMOTE'   
all_reports_df = pd.concat([all_reports_df, report_df.reset_index()], ignore_index=True)

# Confusion matrix
cm = confusion_matrix(y_test, y_pred_class)

# Plotting
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Heatmap for classification report
sns.heatmap(report_df.iloc[:-1, :-1], annot=True, cmap='Blues', fmt='.2f',
            ax=axes[0], linewidths=0.5, linecolor='gray')
axes[0].set_title("Classification Report (Threshold = 0.5)", fontsize=14, fontweight='bold')
axes[0].tick_params(axis='y', rotation=0)

# Confusion matrix plot
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=model.classes_)
disp.plot(cmap='Blues', ax=axes[1], colorbar=False)
axes[1].set_title("Confusion Matrix", fontsize=14, fontweight='bold')
axes[1].grid(False)

plt.tight_layout()
plt.savefig("data/classification_report_and_confusion_matrix_GradientBoosting_SMOTE.png", dpi=300)
plt.show()
```

### GradientBoosting Hyperparameter tuning Classifier
